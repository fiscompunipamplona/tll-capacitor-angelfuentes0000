tomar una placa rectangular de 100x100 y hacer a= 1 (espaciado) aplicar el metodo de super relajacion con gauss seildel y colocar dos lineas de potencial todo para la ecuacion de laplace oviamente en dos dimensiones, ver el comportamiento y la eficiencia del codigo 

```python
from numpy import *#array, linspace,
from math import sin, cos, pi, exp, log
from pylab import plot, xlabel, ylabel, show,imshow,inferno,gray
```

```python

# cosntantes 
M= 100 # 
V= 5.0 # voltaje de la cargas
target= 1e-2 # la tolerancia 
# crear arreglos y evaluar potencial 
phi= zeros([M+1,M+1],float ) # matriz inicial define la caja rectangualar 
rho= zeros([M+1,M+1],float ) # definimos una matriz para las cargas puntuales 
rho[30:80,30:40]=V  # carga uno en la posicion definida
rho[30:80,70:80]=-V # carga dos en la posicion definida
phiprime=empty([M+1,M+1],float ) # matriz donde se guardan los valores nuevos 


```

```python
#print(rho)
```

```python
# main loop
a=1 # espaciado
delta=1 
omega=0.9 # omega cualquiera menor a 1
while delta> target:
    delta=0
    for i in range (1,M): 
        for j in range (1,M):
            diferencia=(phi[i+a,j]+phi[i-a,j]+phi[i,j+a]+phi[i,j-a])/4 -phi[i,j]+rho[i,j]
            phi[i,j]=phi[i,j]+(1+omega)*diferencia# nueva ecuacion con metodo de g-s-sr
            
            if diferencia>delta : delta= diferencia # toma el valor mas alto y compara

imshow(phi) # grafica 
inferno()

show()
           
```
